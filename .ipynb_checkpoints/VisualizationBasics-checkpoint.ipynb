### Correlação das variáveis categóricas

Crammer's V

```python
def cramer_v( var1, var2 ):
    cm = pd.crosstab( var1, var2 ).to_numpy()
    n = cm.sum()
    r, k = cm.shape
    
    chi2 = ss.chi2_contingency( cm )[0]
    chi2corr = max(0, (chi2/n) - (k-1)*(r-1)/(n-1) )
    
    kcorr = k - (k-1)**2/(n-1)
    rcorr = r - (r-1)**2/(n-1)
    
    return np.sqrt( chi2corr / min( kcorr-1, rcorr-1 ) )
```

```python
# Get all categorical data
a = df[['categoria', 'categoria2', 'categoria4']]

# Calculate Cramer V for all combination of categorical data
a1 = cramer_v( a['categoria'], a['categoria'] )
a2 = cramer_v( a['categoria'], a['categoria2'] )
a3 = cramer_v( a['categoria'], a['categoria4'] )

a4 = cramer_v( a['categoria2'], a['categoria'] )
a5 = cramer_v( a['categoria2'], a['categoria2'] )
a6 = cramer_v( a['categoria2'], a['categoria4'] )

a7 = cramer_v( a['categoria4'], a['categoria'] )
a8 = cramer_v( a['categoria4'], a['categoria2'] )
a9 = cramer_v( a['categoria4'], a['categoria4'] )

tab = pd.DataFrame({  
                'categoria': [a1,a2,a3],
                'categoria2': [a4,a5,a6],
                'categoria4': [a7,a8,a9]})
tab.set_index(tab.columns)
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(tab.corr(method='pearson'), annot=True)
```
